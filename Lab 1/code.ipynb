##### **BTVN 2:** Tạo một tensor ban đầu có giá trị là 2. Định nghĩa hàm số và tính gradient
            y = x ^ 3 + 2x ^ 2 + 5x + 1             

- *Hãy tính dy/dx tại giá trị của x*

- *Dùng phương pháp Gradient Descent với Learning Rate Alpha = 0.1 để cập nhật giá trị x trong 10 vòng lặp*


````python
import torch

x = torch.tensor(2.0, requires_grad=True)
y = x**3 + 2*x**2 + 5*x + 1
y.backward()

print("Giá trị x ban đầu =", x.item())
print("Gradient dy/dx tại x =", x.grad.item())
````

````python
alpha = 0.1
x = torch.tensor(2.0, requires_grad=True)

print("Tiến trình của Gradient Descent: ")
print("-" * 40)

for step in range(10):
    if x.grad is not None:
        x.grad.zero_()

    y = x**3 + 2*x**2 + 5*x + 1
    y.backward()
    with torch.no_grad():
        x -= alpha * x.grad
    
    print(f"Vòng {step+1:2d}: x = {x.item():.6f}, dy/dx = {x.grad.item():.6f}")
````

##### **BTVN 3:** Tạo một tập dữ liệu giả lập với x là số giờ học (ngẫu nhiên từ 1 - 10) và y là số điểm tính theo công thức: 
        
            y = 3x + 5 + noise             

**Ghi chú:** __*noise*__ là một giá trị ngẫu nhiên nhỏ

1. Khởi tạo tham số w và b ngẫu nhiên với requires_grad = True
2. Tính MSE
3. Tính Gradient
4. Cập nhật tham số w và b bằng Gradient Descent với learning rate alpha = 0.01
5. Lặp lại quá trình trên trong 100 vòng lặp để xem sự hội tụ của mô hình.

````python
import torch

torch.manual_seed(42)

x = torch.rand(50, 1) * 9 + 1 
noise = torch.randn(50, 1) * 0.5 
y = 3 * x + 5 + noise
````

````python
w = torch.randn(1, requires_grad=True)
b = torch.randn(1, requires_grad=True)
````

````python
def predict(x):
    return w * x + b

def mse_loss(y_pred, y_true):
    return torch.mean((y_pred - y_true) ** 2)
````

````python
alpha = 0.01
````

````python
print("Huấn luyện mô hình Linear Regression")
print("-" * 50)

for epoch in range(100):
    y_pred = predict(x)
    loss = mse_loss(y_pred, y)

    if w.grad is not None:
        w.grad.zero_()
    if b.grad is not None:
        b.grad.zero_()

    loss.backward()

    with torch.no_grad():
        w -= alpha * w.grad
        b -= alpha * b.grad

    if (epoch + 1) % 10 == 0:
        print(f"Epoch {epoch+1:3d}: Loss = {loss.item():.4f}, w = {w.item():.4f}, b = {b.item():.4f}")
````
