## Cho ma trận = [[99,99,99],
##                [99,99,99],
##                [99,99,99]]

### hãy tưởng tượng ô 99 trống 
### giả sử 1 là x 
### giả sử 0 là o

### lấy đầu vào từ phía X và phía O luân phiên 
### nếu đầu vào từ phía x là (0,0)

### ma trận sẽ trở thành 
## [[x,99,99],
##  [99,99,99],
##  [99,99,99]]

# nếu đầu vào từ phía O cũng là (0,0)
# thông báo nhập lại 
# nếu không điền vào ma trận 

# thử thách: Nếu người đầu tiên có 3 phần tử liên tiếp thì dừng trò chơi.



```python
board = [[99, 99, 99], [99, 99, 99], [99, 99, 99]]
player = 'X'

def show():
    for row in board:
        print([cell if cell == 99 else ('X' if cell == 1 else 'O') for cell in row])

def check_win():
    for i in range(3):
        if board[i][0] == board[i][1] == board[i][2] != 99: return True
        if board[0][i] == board[1][i] == board[2][i] != 99: return True
    if board[0][0] == board[1][1] == board[2][2] != 99: return True
    if board[0][2] == board[1][1] == board[2][0] != 99: return True
    return False

while True:
    show()
    try:
        r, c = map(int, input(f"{player} nhập (dòng,cột): ").split(','))
        if board[r][c] != 99:
            print("Ô đã có, nhập lại!")
            continue
        board[r][c] = 1 if player == 'X' else 0
        
        if check_win():
            show()
            print(f"{player} thắng!")
            break
            
        player = 'O' if player == 'X' else 'X'
    except:
        print("Nhập sai, ví dụ: 0,1")
```

```python
# BTVN 4
# Tạo 1 mảng numpy có kích thước 150 x 5 . Giả sử mảng này chứa 150 mảng dữ liệu 
# của nhân viên bao gồm chiều cao , cân nặng , tuổi , lương và điểm trung bình
# Tách mảng dựa trên 4 cột đầu tiên thành 1 biến có tên là x và cột cuối cùng thành y

# Tách mảng x thành x_train và x_test với trong đó x_train chứa 70% dữ liệu và tách y thành y_train và y_test chứa 70% dữ liệu

# Tạo 10 tập dữ liệu x_train không chồng chéo nhau
  
```

```python
import numpy as np

# Tạo mảng 150 x 5
np.random.seed(42)  # Để kết quả có thể tái tạo
data = np.random.randn(150, 5)
print(f"Kích thước mảng: {data.shape}")
print("5 dòng đầu tiên của mảng:")
print(data[:5])
print("-" * 50)
```

```python
# Tách thành x (4 cột đầu) và y (cột cuối)
x = data[:, :4]
y = data[:, 4]
print(f"Kích thước x: {x.shape}")
print(f"Kích thước y: {y.shape}")
print("x (5 dòng đầu):")
print(x[:5])
print("y (5 giá trị đầu):")
print(y[:5])
print("-" * 50)
```

```python
# Tách mảng dựa trên 4 cột đầu tiên thành 1 biến có tên là x và cột cuối cùng thành y
n_samples = len(x)
n_train = int(0.7 * n_samples)
x_train = x[:n_train]
x_test = x[n_train:]
y_train = y[:n_train]
y_test = y[n_train:]

print(f"Số mẫu train: {n_train}, số mẫu test: {n_samples - n_train}")
print(f"x_train shape: {x_train.shape}")
print(f"x_test shape: {x_test.shape}")
print(f"y_train shape: {y_train.shape}")
print(f"y_test shape: {y_test.shape}")
print("-" * 50)
```

```python
#Tạo 10 tập dữ liệu x_train không chồng chéo nhau
n_folds = 10
fold_size = n_train // n_folds
print(f"Kích thước mỗi fold: {fold_size} mẫu")

x_folds = []
for i in range(n_folds):
    start_idx = i * fold_size
    end_idx = start_idx + fold_size
    x_fold = x_train[start_idx:end_idx]
    x_folds.append(x_fold)
    print(f"Fold {i+1}: chỉ số {start_idx}-{end_idx-1}, shape: {x_fold.shape}")

print("\nKiểm tra không chồng chéo:")
total_samples = sum([fold.shape[0] for fold in x_folds])
print(f"Tổng số mẫu trong 10 folds: {total_samples}")
print(f"Số mẫu x_train: {x_train.shape[0]}")
print(f"Kiểm tra: {total_samples} == {x_train.shape[0]}? {total_samples == x_train.shape[0]}")
```
